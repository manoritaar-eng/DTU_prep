```python
import numpy as np 
import matplotlib.pyplot as plt
import scipy.stats as stats

x = np.random.normal(loc=0, scale=1, size=1000)

plt.hist(x, bins = 30, density = True, alpha = 0.5, color = "blue")

xs = np.linspace(-4,4,100000)
plt.plot(xs, stats.norm.pdf(xs, 0, 1))
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Histogram of Random Samples from Normal Distribution')
plt.show()

#Exponential distribution (x>=0)
y = np.random.exponential(scale=1, size=1000)
plt.hist(y, bins = 30, density = True, alpha = 0.5, color = "green")

ys = np.linspace(0,4,200)
plt.plot(ys, stats.expon.pdf(ys, 0, 1))
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Histogram of Random Samples from Exponential Distribution')
plt.show()

#Uniform distribution (0<=x<=1)

z = np.random.uniform(low=0, high=1, size=1000)
plt.hist(z, bins = 30, density = True, alpha = 0.5, color = "red")

zs = np.linspace(0,1,200)
plt.plot(zs, stats.uniform.pdf(zs, 0, 1))
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Histogram of Random Samples from Uniform Distribution')      
plt.show()

#Binomial distribution - Discrete - np.arange()

b = np.random.binomial(n=10, p=0.5, size=1000)
plt.hist(b, bins = 30, density = True, alpha = 0.5, color = "orange")

bs = np.arange(0,11)
plt.plot(bs, stats.binom.pmf(bs, n=10, p=0.5))
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Histogram of Random Samples from Binomial Distribution')
plt.show()  

#Poisson distribution - Discrete - np.arange()

c = np.random.poisson(lam=3, size=1000)
plt.hist(c, bins = 30, density = True, alpha = 0.5, color = "purple")

cs = np.arange(0,12)
plt.plot(cs, stats.poisson.pmf(cs, mu=3))
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Histogram of Random Samples from Poisson Distribution')
plt.show()
```
