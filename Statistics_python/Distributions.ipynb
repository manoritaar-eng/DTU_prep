```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt

#print(stats.poisson.cdf(4,3))

mean = 54
variance = 16
std_dev = 3

#P(X<=20)

print(stats.norm.cdf(10, mean, std_dev))

#print(stats.norm.ppf(stats.norm.cdf(2)))

x = np.linspace(-4,4,1000)
cdf_norm = (stats.norm.cdf(2,1,1))
#print(cdf_norm)

plt.figure(figsize=(8,5))
plt.plot(x, stats.norm.cdf(x, loc = 1, scale = 2), label = 'CDF', color = 'blue')
plt.title('*****CDF Normal distribution*****')
plt.xlabel('x')
plt.ylabel('Cumulative Probability')
plt.grid(True)
plt.show()


#Binomial distribution parameters
n = 20
p = 0.8

k = np.arange(0, n+1)
prob = stats.binom.pmf(k,n,p)

plt.figure(figsize = (8,5))
plt.bar(k, prob, color = 'blue')
plt.title(f'Binomial distribution PMF (n={n}, p={p})')
plt.xlabel('Number of successes')
plt.ylabel('Probability')
plt.show()

#Hypergeometric distribution parameters
M = 20
n = 12 
N = 5 

k = np.arange(0, N+1)

pmf_hypergeom = stats.hypergeom.pmf(k, M, n, N)
cdf_hypergeom = stats.hypergeom.cdf(k, M, n, N)

plt.figure(figsize = (10,5))

plt.bar(k, pmf_hypergeom, color = 'green', alpha = 0.7)
plt.title(f'Hypergeometric distribution PMF (M={M}, n={n}, N={N})')
plt.xlabel('Number of successes k')
plt.ylabel('Probability')
plt.grid(True)
plt.show()

plt.plot(k, cdf_hypergeom, label = 'CDF', color = 'purple')
plt.title('CDF Hypergeometric distribution')
plt.xlabel('Number of successes k')
plt.ylabel('Cumulative Probability')
plt.grid(True)
plt.tight_layout()
plt.show()




#Calculate PDF and CDF of Normal distribution

x = np.linspace(-4, 4, 1000)

cdf_1 = stats.norm.cdf(10, loc = 65, scale = 4)
cdf_2 = stats.norm.cdf(10, loc = 54, scale = 3)

plt.figure(figsize=(10,5))

#Plot PDF
plt.subplot(1,2,1)
plt.plot(x, cdf_1, label = 'CDF', color = 'red')
plt.title('CDF 1 Normal distribution')
plt.ylabel('Cumulative Probability')
plt.grid(True)

#Plot CDF 
plt.subplot(1,2,2)
plt.plot(x, cdf_2, label = 'CDF', color = 'blue')
plt.title('CDF 2 Normal distribution')
plt.ylabel('Cumulative Probability')
plt.grid(True)

plt.tight_layout()
plt.show()








```
