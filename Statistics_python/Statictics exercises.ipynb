```python
import numpy as np
import pandas as pd
#import scipy.stats as stats
import matplotlib.pyplot as plt
#import statsmodels.api as sm
#import statsmodels.stats.proportion as smprop

#I have used a pandas data frame to do this task but a Numpy array can also be used. 
# Infant Birth weight  

data = pd.DataFrame({
    'females' : [2474, 2547, 2830, 3219, 3429, 3448, 3677, 3872, 4001, 4111],
    'males' : [2844, 2863, 2963, 3239, 3379, 3449, 3582, 3926, 4151, 4356]
})

mean_females = data['females'].mean()
variance_females = data['females'].var()
std_dev_females = data['females'].std()
coefficient_of_variation_f = (std_dev_females / mean_females)*100

print("Mean (Females):", mean_females)
print("Variance (Females):", variance_females)
print("Standard Deviation (Females):", std_dev_females)
print(f"Coefficient of Variation (Females): {coefficient_of_variation_f}%")

mean_males = data['males'].mean()
variance_males = data['males'].var()
std_dev_males = data['males'].std()     
coefficient_of_variation_m = (std_dev_males / mean_males)*100

print("Mean (Males):", mean_males)
print("Variance (Males):", variance_males)
print("Standard Deviation (Males):", std_dev_males)
print(f"Coefficient of Variation (Males): {coefficient_of_variation_m}%")

min_quartile_females = data['females'].quantile(0)
first_quartile_females = data['females'].quantile(0.25)
middle_quartile_females = data['females'].quantile(0.5)
third_quartile_females = data['females'].quantile(0.75)
max_quartile_females = data['females'].quantile(1)

print("0th Percentile (Females):", min_quartile_females)
print("25th Percentile (Females):", first_quartile_females)
print("50th Percentile (Females):", middle_quartile_females)
print("75th Percentile (Females):", third_quartile_females)
print("100th Percentile (Females):", max_quartile_females)

min_quartile_males = data['males'].quantile(0)
first_quartile_males = data['males'].quantile(0.25)
middle_quartile_males = data['males'].quantile(0.5)
third_quartile_males = data['males'].quantile(0.75)
max_quartile_males = data['males'].quantile(1)

print("0th Percentile (Males):", min_quartile_males)
print("25th Percentile (Males):", first_quartile_males)
print("50th Percentile (Males):", middle_quartile_males)
print("75th Percentile (Males):", third_quartile_males)
print("100th Percentile (Males):", max_quartile_males)

#plt.boxplot([data['females'], data['males']])

#plt.xlabel('Females')
#plt.ylabel('Values')
#plt.title('Boxplot Modified')
#plt.show()

#plt.boxplot(data['males'])
#plt.xlabel('Males')
#plt.ylabel('Values')
#plt.title('Boxplot of Male Data')
#plt.show()

#Cholesterol 

x = np.array([9.1, 8.0, 7.7, 10.0, 9.6, 7.9, 9.0, 7.1, 8.3, 9.6, 8.2, 9.2, 7.3, 8.5, 9.5])
y = np.array([8.2, 6.4, 6.6, 8.5, 8.0, 5.8, 7.8, 7.2, 6.7, 9.8, 7.1, 7.7, 6.0, 6.6, 8.4])

median_before = np.median(x)
median_after = np.median(y)

print("Median before:", median_before)
print("Median after:", median_after)

std_dev_before = np.std(x, ddof=1)  #ddof = 1 means the denominator is n-1 which more accurate than ddof=0 where denominator is n.
std_dev_after = np.std(y, ddof=1)

print("Standard Deviation before:", std_dev_before)
print("Standard Deviation after:", std_dev_after)

data = np.stack((x, y))
covariance = np.cov(data)

print("Covariance:", covariance)

Correlation_coefficient = np.corrcoef(data)
print("Correlation Coefficient:", Correlation_coefficient)

Diff = y-x
quartile_diff = np.quantile(Diff, [0,0.25,0.5,0.75,1], method = 'averaged_inverted_cdf')
print("Quartiles of the differences:", quartile_diff)

print("Mean:",np.mean(Diff))
print("Variance:",np.var(Diff, ddof = 1))
print(np.std(Diff, ddof = 1))

fig = plt.boxplot(Diff)
plt.show()


           
```
